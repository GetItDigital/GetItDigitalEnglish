```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import math as mt
```

Resonanzkreis eines
RLC in Reihe

```python
R=50 #Werte vernderbar
L=0.01 #Werte vernderbar
C=0.00001 #Werte vernderbar
W0=1/mt.sqrt(L*C)
print("W0=",W0)
wp = np.linspace(W0-W0+1,W0+W0,100)
w=wp/W0
Xc = 1/(wp*C) #Betrag
Xl= wp*L #Betrag
X=abs(wp*L-1/(wp*C))
Z=abs(R+wp*L-1/(wp*C))
plt.ylim(0,R+R)
plt.plot(w, Xc, "r", label="Xc")
plt.legend()
plt.plot(w, Xl, "g", label="XL")
plt.legend()
plt.plot(w, X, "y", label="X")
plt.legend()
plt.plot(w, Z, "b", label="Z")
plt.legend()
plt.xlabel('W/W0')
plt.ylabel('|Z|')


import tikzplotlib

tikzplotlib.save("ortskurve.tex")
```

Man sieht, dass sich bei der Grenzfrequenz W0 sich Xl und Xc aufheben.
