### Aufgabe 1

Gegeben sind die Spannungen   
 
$\quad u_1(t) = 5 V ·sin(2 π·50 Hz ·t + 60◦)$  
$\quad u_2(t) = 8 V ·sin(2 π·50 Hz ·t −10◦)$  
  
Berechnen Sie die aus der Addition der beiden Spannungen resultierende Spannung
      
$\quad u_{res}(t) = u_1(t) + u_2(t)$   
 
Skizzieren Sie als Hilfe ein Zeigerdiagramm.  

```python
# Bibliotheken importieren (Beim Start immer erst ausführen)
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Winkel in Radiant umrechnen
phi1 = 60*np.pi/180
phi2 = -10*np.pi/180
t=0
u1 = 5*np.sin(2*np.pi*50*t+ phi1)
u2 = 8*np.sin(2*np.pi*50*t+ phi2)

# Spannungen in die komplexe Ebene überführen

u1k = 5*np.exp(1j*phi1)
u2k = 8*np.exp(1j*phi2)

# Real- und Imaginärteil der Spannungen berechnen
u1r = np.real(u1k)
u1i = np.imag(u1k)
u2r = np.real(u2k)
u2i = np.imag(u2k)

# Spannungen addieren Realteil und Imaginärteil seperat
uresr = u1r + u2r
uresi= u1i + u2i

uresbetrag= np.abs(uresr + 1j*uresi)
uresphase = np.angle(uresr + 1j*uresi)
print('Ures =', uresbetrag, 'V * sin(2*pi*50*t +', uresphase*180/np.pi, '°)')


# Plot
fig, ax = plt.subplots()
# u1
ax.arrow(0, 0, u1r, u1i, width=0.05, length_includes_head = True, head_width= 5*0.05,color='r', label='${U}_1$')
# u1 addieren
ax.arrow(u2r, u2i, u1r, u1i, width=0.05, length_includes_head=True, head_width= 5*0.05,color='r', label=r'\underline{U}_2$',linestyle='dotted')
# u2 
ax.arrow(0, 0, u2r, u2i, width=0.05, length_includes_head=True, head_width= 5*0.05,color='b', label=r'\underline{U}_2$')
# u2 addieren
ax.arrow(u1r, u1i, u2r, u2i, width=0.05, length_includes_head=True, head_width= 5*0.05,color='b', label=r'\underline{U}_2$',linestyle='dotted')
# ures
ax.arrow(0, 0, uresr, uresi, width=0.05, length_includes_head=True, head_width= 5*0.05,color='black', label=r'\underline{U}_{res}$')

plt.grid()
plt.xlim(-uresbetrag, uresbetrag)
plt.ylim(-uresbetrag, uresbetrag)

plt.title('Zeigerdiagramm', fontsize=16)
plt.ylabel('Im', loc='top', fontsize=16, rotation=90)
plt.xlabel('Re', loc='right', fontsize=16)




# Achsenverschiebung
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Plotgröße
plt.rcParams['figure.figsize'] = [8, 8]
```

```python
UL= 400
f = 50
R= 50
XL = 100
# a)

# Betrag und Phase des Widerstandes
Rbetrag = np.sqrt(R**2 + XL**2)
Rphi = np.arctan(XL/R)*180/np.pi
R= Rbetrag * np.exp(1j*Rphi*np.pi/180)
print('Z =', Rbetrag, 'Ohm * e^', Rphi, '°')

# Spannungen im Stern
u1s = UL/np.sqrt(3) * np.exp(1j*0)
u2s = UL/np.sqrt(3) * np.exp(1j*(-120*np.pi/180))
u3s = UL/np.sqrt(3) * np.exp(1j*(120*np.pi/180))
print('U1 im Stern =', u1s)
print('U2 im Stern =', u2s)
print('U3 im Stern =', u3s)

# Ströme im Stern
i1s = u1s/R
i2s = u2s/R
i3s = u3s/R
print('I1 im Stern =', i1s)
print('I2 im Stern =', i2s)
print('I3 im Stern =', i3s)

# Spannungen im Dreieck
u12d = UL * np.exp(1j*0)
u23d = UL * np.exp(1j*(-120*np.pi/180))
u31d = UL * np.exp(1j*(120*np.pi/180)) 
print('U12 im Dreieck =', u12d)
print('U23 im Dreieck =', u23d)
print('U31 im Dreieck =', u31d)

# Ströme im Dreieck
i12d = u12d/R
i23d = u23d/R
i31d = u31d/R
i1d = i12d - i31d
i2d = i23d - i12d
i3d = i31d - i23d
print('I12 im Dreieck =', i12d)
print('I23 im Dreieck =', i23d)
print('I31 im Dreieck =', i31d)
print('I1 im Dreieck =', i1d)
print('I2 im Dreieck =', i2d)
print('I3 im Dreieck =', i3d)

# Real- und Imaginärteil der Spannungen und Ströme berechnen im Stern
u1sr = np.real(u1s)
u1si = np.imag(u1s)
u2sr = np.real(u2s)
u2si = np.imag(u2s)
u3sr = np.real(u3s)
u3si = np.imag(u3s)

# Verkettete Spannungen berechnen im Stern
u12sr = u2sr - u1sr
u12si = u2si - u1si
u23sr = u3sr - u2sr
u23si = u3si - u2si
u31sr = u1sr - u3sr
u31si = u1si - u3si

# Real- und Imaginärteil der Ströme berechnen im Stern(100fach zur Visualisierung)
i1sr = np.real(i1s)*100
i1si = np.imag(i1s)*100
i2sr = np.real(i2s)*100
i2si = np.imag(i2s)*100
i3sr = np.real(i3s)*100
i3si = np.imag(i3s)*100

# Real- und Imaginärteil der Ströme berechnen im Dreieck(100fach zur Visualisierung)
i12dr = np.real(i12d)*100
i12di = np.imag(i12d)*100
i23dr = np.real(i23d)*100
i23di = np.imag(i23d)*100
i31dr = np.real(i31d)*100
i31di = np.imag(i31d)*100

# b)
# Plot im Stern
fig, ax1 = plt.subplots()

# u1
ax1.arrow(0, 0, u1sr,u1si, width=1, length_includes_head = True, head_width= 7.5*1,color='r', label='U1')
# u2 
ax1.arrow(0, 0, u2sr, u2si, width=1, length_includes_head=True, head_width= 7.5*1,color='r', label='U2')
# u3
ax1.arrow(0, 0, u3sr, u3si, width=1, length_includes_head=True, head_width= 7.5*1,color='r', label='U3')

# u12
ax1.arrow(u1sr, u1si, u12sr, u12si, width=1, length_includes_head=True, head_width= 7.5*1,color='r', label='U12')
# u23
ax1.arrow(u2sr, u2si, u23sr, u23si, width=1, length_includes_head=True, head_width= 7.5*1,color='r', label='U23')
# u31
ax1.arrow(u3sr, u3si, u31sr, u31si, width=1, length_includes_head=True, head_width= 7.5*1,color='r', label='U31')

# i1
ax1.arrow(0, 0, i1sr, i1si, width=1, length_includes_head = True, head_width= 7.5*1,color='b', label='I1')
# i2
ax1.arrow(0, 0, i2sr, i2si, width=1, length_includes_head=True, head_width= 7.5*1,color='b', label='I2')
# i3
ax1.arrow(0, 0, i3sr, i3si, width=1, length_includes_head=True, head_width= 7.5*1,color='b', label='I3')

plt.title('Zeigerdiagramm im Stern', fontsize=16)
plt.ylabel('Im', loc='top', fontsize=16, rotation=90)
plt.xlabel('Re', loc='right', fontsize=16)

# Achsenverschiebung
ax1.spines['left'].set_position('zero')
ax1.spines['bottom'].set_position('zero')

ax1.set_xticklabels([])
ax1.set_yticklabels([])

#Plot im Dreieck
fig, ax2 = plt.subplots()

# u12
ax2.arrow(0, 0, u12d.real, u12d.imag, width=4, length_includes_head = True, head_width= 7.5*2,color='r', label='U12')
# u23
ax2.arrow(0, 0, u23d.real, u23d.imag, width=4, length_includes_head=True, head_width= 7.5*2,color='r', label='U23')
# u31
ax2.arrow(0, 0, u31d.real, u31d.imag, width=4, length_includes_head=True, head_width= 7.5*2,color='r', label='U31')

# i1
ax2.arrow(0, 0, i12dr, i12di, width=4, length_includes_head = True, head_width= 7.5*2,color='b', label='I1')
# i2
ax2.arrow(0, 0, i23dr, i23di, width=4, length_includes_head=True, head_width= 7.5*2,color='b', label='I2')
# i3
ax2.arrow(0, 0, i31dr, i31dr, width=4, length_includes_head=True, head_width= 7.5*2,color='b', label='I3')

plt.title('Zeigerdiagramm im Dreieck', fontsize=16)
plt.ylabel('Im', loc='top', fontsize=16, rotation=90)
plt.xlabel('Re', loc='right', fontsize=16)


# Achsenverschiebung
ax2.spines['left'].set_position('zero')
ax2.spines['bottom'].set_position('zero')

ax2.set_xticklabels([])
ax2.set_yticklabels([])

# Plotgröße
plt.rcParams['figure.figsize'] = [8, 8]

# c)
# Leistungsfaktor berechnen
cosphi = np.cos(Rphi*np.pi/180)
print('cos(phi) =', cosphi)

```
